```python
def bytePairEncode(s):
    new_token_ord = 90
    encoding = {}
    new_string = s
    
    while True:
        s = new_string
        pairs = []
        for i in range(1,len(s)):
            pair = (s[i-1],s[i])
            pairs.append(pair)
        freq = {}
        for p in pairs:
            freq[p] = freq.get(p,0)+1
        max_freq = max(freq.values())
        if max_freq == 1:        
            break

        max_freq_ind = list(freq.values()).index(max_freq)
        max_freq_key = list(freq.keys())[max_freq_ind]
        encoding[chr(new_token_ord)] = max_freq_key[0]+max_freq_key[1]
        
        new_string = []
        i = 1
        while i < len(s):
            if (s[i-1],s[i]) == max_freq_key:
                new_string.append(chr(new_token_ord))
                i += 2
            else:
                new_string.append(s[i-1])
                i += 1
        if i == len(s):
            new_string.append(s[i-1])
        new_string = ''.join(new_string)
        new_token_ord -= 1
        print(new_string, encoding)

s = 'aaabdaaabac'
bytePairEncode(s)
```

```python
def bytePairEncode(s):
    new_token_ord = 90
    encoding = {}
    new_string = list(s.encode('utf-8'))
    
    while True:
        s = new_string
        freq = {}
        for p in zip(s,s[1:]):
            freq[p] = freq.get(p,0)+1
        max_freq = max(freq.values())
        if max_freq == 1:        
            break

        max_freq_ind = list(freq.values()).index(max_freq)
        max_freq_key = list(freq.keys())[max_freq_ind]
        encoding[new_token_ord] = (max_freq_key[0],max_freq_key[1])
        
        new_string = []
        i = 1
        while i < len(s):
            if (s[i-1],s[i]) == max_freq_key:
                new_string.append(new_token_ord)
                i += 2
            else:
                new_string.append(s[i-1])
                i += 1
        if i == len(s):
            new_string.append(s[i-1])
        new_token_ord -= 1
        print(new_string, encoding)
    return new_string, encoding

s = 'aaabdaaabac'
ns, enc = bytePairEncode(s)
print(''.join([chr(i) for i in ns]))
print({chr(k): chr(v[0])+chr(v[1]) for k,v in enc.items()})
```
